```python
names=['Ramesh','Suresh','Sathish']
age=[30,35,40]
df=pd.DataFrame(zip(names,age),
                 columns=['Name','Age'],
                  index=['A','B','C'])
city_names=['Hyd','blr','Chennai']
df['City']=city_names

df.drop('city',
         axis=0,
          inplace=True)
df
```
