# One hot encoder

-one hot encder name says at a time one will on and other will off

-For exmple case status has two lables

   *Certified
   
   *Denied
   
-when you apply ne hot encoding on case status,it creates two more extra columns

   *case_statuss_Certified
   
   *case_status_Denied
   
Advantages:

-when you develop ML model it is very important the columns should be independent each other

-so here case status creating two extra column

-which are independent each other,which means the row values at a time only one columns has.

-Columns are independent each other

-90 degrees to each other

-perpendicular to each other

-Orthogonal to each other

Disadvantags:

-If a columns has 100 unique lables 10 new columns will be created.

-Columns are more means,Dimensions are more.

-Data set will become sparse,which means huge

-The processing time is more

-The memory consumption is more

-This is called Curse of Dimenisionality
